#### creating the list of all dates from max to min (which is take from the dataframe present)

```python
def generate_date_list(start_date, end_date):

    start = datetime.strptime(start_date, '%Y-%m-%d')
    end = datetime.strptime(end_date, '%Y-%m-%d')
    date_list = []
    current_date = start
    while current_date <= end:
        date_list.append(current_date.strftime('%Y-%m-%d'))
        current_date += timedelta(days=1)
    return date_list
```

```python
start_year = str(pd.to_datetime(df['DATE']).dt.year.min())
```

```python
start_date = f'{start_year}-01-01'
start_date
end_date = str(df['DATE'].max())
dates = generate_date_list(start_date, end_date)
```

```python
date_df = pd.DataFrame(dates,columns=['DATE'])
```

```python
date_df['MONTH'] = pd.to_datetime(date_df['DATE']).dt.month_name()
date_df['YEAR'] = pd.to_datetime(date_df['DATE']).dt.year.astype('str')
date_df['MONTH_YEAR'] = date_df['MONTH'] +'_'+date_df['YEAR']
date_df['DAY'] = pd.to_datetime(date_df['DATE']).dt.day_name()
```

#### after sorting saturday with dates add the sequence number so we can take 2 and 4 saturday which are present for the months

```python
saturday_df = date_df[date_df['is_saturday']==True][['MONTH_YEAR','DATE']].sort_values('DATE')
saturday_df['sequence'] = date_df[date_df['is_saturday']==True].groupby('MONTH_YEAR').cumcount() + 1
```

```python
saturday_df
```

```python
saturday_df[saturday_df['sequence']==2]
```

#### Finding the second saturday from csv

```python
df.loc[df['DATE'].isin(saturday_df[saturday_df['sequence']==2]['DATE'].unique()),'is_saturday2']=True
```

#### Finding the fourth saturday from csv

```python
df.loc[df['DATE'].isin(saturday_df[saturday_df['sequence']==4]['DATE'].unique()),'is_saturday4']=True
```
